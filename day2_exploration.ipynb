```python
import pandas as pd
import glob

files = glob.glob( "data/raw/play_by_play_*.parquet")

pbp_all = pd.concat ([pd.read_parquet(f) for f in files], ignore_index=True)

print( "Combined dataset shape:", pbp_all.shape)
pbp_all.head()
```

```python
print( "Rows, Columns:", pbp_all.shape)

pbp_all.head()

pbp_all.tail()
```

```python
print(pbp_all.columns.tolist())

pbp_all.info()

pbp_all.describe()
```

```python
print(pbp_all['pass'].value_counts())
print(pbp_all['rush'].value_counts())

print( "Average yards gained:", pbp_all['yards_gained'].mean())
```

```python
team_yards = pbp_all.groupby( "posteam")["yards_gained"].mean().sort_values(ascending=False)
print(team_yards)

team_plays = pbp_all["posteam"].value_counts()
print(team_plays)
```

```python
team_yards = pbp_all.groupby( "posteam")["yards_gained"].mean().sort_values(ascending=False)
print( "Average yards gained per team:")
print(team_yards)

team_plays = pbp_all["posteam"].value_counts()
print( "\nTotal number of plays per team:")
print(team_plays)
```

```python
season_yards = pbp_all.groupby( "season")["yards_gained"].mean()
print( "Average yards gained per season:")
print(season_yards)

season_plays = pbp_all["season"].value_counts().sort_index()
print( "\nTotal number of plays per season:")
print(season_plays)
```

```python
play_type_yards = pbp_all.groupby( "play_type")["yards_gained"].mean().sort_values(ascending=False)
print("Average yards gained by play type: ")
print(play_type_yards)

play_type_counts = pbp_all["play_type"].value_counts()
print( "\nNumber of plays by type:")
print(play_type_counts)
```
