```python
import pandas as pd
import numpy as np
import glob, os
```

```python
files = glob.glob( "data/raw/play_by_play_*.parquet")
print( "Files found:", len(files))
print(files[:5])
```

```python
pbp_all = pd.concat([pd.read_parquet(f) for f in files], ignore_index=True)
print( "Rows, Cols:", pbp_all.shape)
pbp_all.head()
```

```python
pbp_all["game_key"] = pbp_all["game_id"]
pbp_all["season"]  = pd.to_numeric(pbp_all["game_id"].str[:4], errors="coerce")

pbp = pbp_all.loc[pbp_all["season_type"] == "REG"].copy()

is_pass_play = pbp["pass"] == 1
is_rush_play = pbp["rush"] == 1
len(pbp),  is_pass_play.sum(), is_rush_play.sum()
```

```python
qb_games = (
    pbp[is_pass_play]
    .groupby(["passer_player_name", "game_id"])
    .agg(
        attempts=("play_id", "count"),
        completions=("complete_pass", "sum"),
        pass_yards=("yards_gained", "sum"),
        pass_tds=("pass_touchdown", "sum"),
        interceptions=("interception", "sum")
        )
        .reset_index()
)

qb_games["comp_pct"] = qb_games["completions"] / qb_games["attempts"] * 100

qb_games.head()
```

```python
qb_games.groupby("passer_player_name")["attempts"].sum().sort_values(ascending=False).head(10)
```

```python
defense_stats = (
    pbp.groupby(["season", "defteam"])
    .agg(
        pass_def_epa=( "epa", lambda x: x[pbp["pass"] == 1].mean()),
        rush_def_epa=( "epa", lambda x: x[pbp["rush"] == 1].mean())
    )
    .reset_index()
)

defense_stats.head()
```

```python
qb_games = (
    pbp[is_pass_play & pbp["passer_player_name"].notna()]
    .groupby(["season", "week", "game_id", "posteam", "defteam", "passer_player_name"])
    .agg(
        pass_yards=("yards_gained","sum"),
        attempts=("play_id","count"),
        completions=("complete_pass","sum")
    )
    .reset_index()
)

qb_games = qb_games.merge(
    defense_stats,
    on=["season","defteam"],
    how="left"
)

qb_games.head()
```

```python
rb_games = (
    pbp[is_rush_play & pbp["rusher_player_name"].notna()]
    .groupby(["season","week","game_id","posteam","defteam","rusher_player_name"])
    .agg(
        rush_yards=("yards_gained","sum"),
        rush_att=("play_id","count")
    )
    .reset_index()
)

rb_games = rb_games.merge(
    defense_stats,
    on=["season","defteam"],
    how="left"
)

rb_games.head()
```

```python
qb_games = qb_games.sort_values(["passer_player_name","season","week"])

qb_games["qb_3g_avg_yards"] = (
    qb_games.groupby("passer_player_name")["pass_yards"]
    .transform(lambda x: x.rolling(3, 1).mean())
)
qb_games["qb_3g_avg_att"] = (
    qb_games.groupby("passer_player_name")["attempts"]
    .transform(lambda x: x.rolling(3, 1).mean())
)
qb_games.head(10)
```

```python
rb_games = rb_games.sort_values(["rusher_player_name","season","week"])

rb_games["rb_3g_avg_yards"] = (
    rb_games.groupby("rusher_player_name")["rush_yards"]
    .transform(lambda x: x.rolling(3, 1).mean())
)
rb_games["rb_3g_avg_att"] = (
    rb_games.groupby("rusher_player_name")["rush_yards"]
    .transform(lambda x: x.rolling (3, 1).mean())
)
rb_games.head(10)
```

```python
qb_games[qb_games["passer_player_name"]=="Patrick Mahomes"][["season","week","pass_yards","qb_3g_avg_yards"]].head(10)

rb_games[rb_games["rusher_player_name"]=="Derrick Henry"][["season","week","rush_yards","rb_3g_avg_yards"]].head(10)
```

```python
import matplotlib.pyplot as plt

mahomes = qb_games[qb_games["passer_player_name"]=="Patrick Mahomes"]

plt.plot(mahomes["week"], mahomes["pass_yards"], label="Pass Yards")
plt.plot(mahomes["week"], mahomes["qb_3g_avg_yards"], label ="3-Game Avg", linewidth=3)
plt.xlabel("Week")
plt.ylabel("Yards")
plt.title("Patrick Mahomes Passing Yards vs 3-Game Avg")
plt.legend()
plt.show()
```

```python
henry = rb_games[rb_games["rusher_player_name"]=="Derrick Henry"]

plt.plot(henry["week"], henry["rush_yards"], label="Rush Yards")
plt.plot(henry["week"], henry["rb_3g_avg_yards"], label="3-Game Avg", linewidth=3)
plt.xlabel("Week")
plt.ylabel("Yards")
plt.title("Derrick Henry Rushing Yards vs 3-Game Avg")
plt.legend()
plt.show()
```
